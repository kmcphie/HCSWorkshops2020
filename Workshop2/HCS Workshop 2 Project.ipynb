```python
# import requests package and set up page

import requests

page = requests.get("https://www.brainyquote.com/topics/inspirational-quotes")
print(page.status_code)
```

```python
# import BeautifulSoup and make a "BeautifulSoup object"

from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')

# print(soup.prettify())
```

```python
# scrape some inspirational quotes

quote_grid = soup.find(id = 'quotesList')
quotes = quote_grid.find_all(class_ = 'clearfix')

for quote in quotes:
    print(quote.get_text())
    print("---------")
```
